```python
import pandas as pd
import joblib
```

```python
data = {
    'Gender': [1],
    'Married': [0],
    'Dependents': ['4'],
    'Education': [1],
    'Self_Employed': [1],
    'ApplicantIncome': [5849],
    'CoapplicantIncome': [2000],
    'LoanAmount': [300],
    'Loan_Amount_Term': [360],
    'Credit_History': [1],
    'Property_Area': [2]
}
df = pd.DataFrame(data)
print(df)
```

```python
scaler = joblib.load("scaler_minmax.joblib")
model = joblib.load("rf_model.joblib")
```

```python
columns_to_normalize = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
# Apply Min-Max scaling to the specified columns
df[columns_to_normalize] = scaler.transform(df[columns_to_normalize])
print(df[columns_to_normalize].head())
```

```python
y_pred = model.predict(df)
print(y_pred)
```
